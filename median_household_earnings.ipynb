```python
def get_avg_residents(residents_2001, residents_2011, residents_2021):

    areas = list(residents_2001.index.values)
    avg_residents_df = pd.DataFrame()

    for area in areas:

        # Initialise DataFrame with census data and missing years in between
        nan_area_df = pd.DataFrame({"1": [residents_2001.loc[area][0], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][0], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][0]],
                           "2": [residents_2001.loc[area][1], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][1], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][1]],
                           "3": [residents_2001.loc[area][2], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][2], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][2]],
                           "4": [residents_2001.loc[area][3], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][3], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][3]],
                           "5": [residents_2001.loc[area][4], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][4], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][4]],
                           "6": [residents_2001.loc[area][5], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][5], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][5]],
                           "7": [residents_2001.loc[area][6], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][6], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][6]],
                           "8": [residents_2001.loc[area][7], None, None, None, None, None, None, None, None, None, residents_2011.loc[area][7], None, None, None, None, None, None, None, None, None, residents_2021.loc[area][7]]},
                           index=list(range(2001,2022)))
        
        # Interpolate missing values
        area_df = nan_area_df.interpolate(method='quadratic', limit_direction ='forward')

        # Create a DataFrame which has the total number of residents
        total_residents_df = area_df.copy()

        for i in range(0,8):
            total_residents_df.iloc[:,i] = total_residents_df.iloc[:,i].apply(lambda x: x * (i+1))

        # Get average number of residents per house by year
        total_residents = total_residents_df.sum(axis=1)
        total_households = area_df.sum(axis=1)
        avg_residents = total_residents.divide(total_households)

        # Append column to DataFrame
        avg_residents_df[area] = avg_residents

        # Transpose DataFrame to get years as columns and area as rows
        
    return avg_residents_df.transpose()
```

```python
average_residents = get_avg_residents(residents_2001, residents_2011, residents_2021)
```
